# Legendre Polynomials $P_l(x)$

```python
from sympy.abc import x,n
from IPython.display import display

print("These are the first five Legendre Polynomials:")

for i in range(5):
    display(sympy.legendre(i,x))
```

```python
def legPol(nb):
    N = 1000
    xvals = np.linspace(-1,1,N)
    for i in range(nb):
        p = legendre(i)
        y = p(xvals)
        plt.plot(xvals,y,label = sympy.legendre(i,x))
```

```python
legPol(5)
plt.legend(bbox_to_anchor=(1, 1))
```

```python
#Using the polar to cartesian transformation x = r*cosT, y = r*sinT
T = np.linspace(0,10*np.pi,1000)

fig1, ax1 = plt.subplots(5,2, figsize=(6, 10),constrained_layout=True)

ra = []
for i in range(4):
    for j in range(i,4):
        ra.append(lpmv(i,j,np.cos(T)))
        
rb = 0
for i in range(5):
    for j in range(2):
        ax1[i][j].plot(ra[rb]*np.cos(T),ra[rb]*np.sin(T))
        rb += 1
```

# Hydrogen Atom Wave Function using Laguerre Polynomials and Assiciated Functions

```python
from sympy import assoc_laguerre as asl
from sympy import Ynm
#assoc_laguerre(l,m,x)

#Calculates the radial wave function for hydrogen
def R_sym(nq,l,r):
    return sympy.sqrt(((2/(nq*a))**3)*(sympy.factorial(nq-l-1)/(2*nq*(sympy.factorial(nq+l))**3)))*sympy.exp(-r/(nq*a))*((2*r/(nq*a))**l)*sympy.factorial(nq+l)*asl(nq-l-1, 2*l+1, 2*r/(nq*a))

def Psi_sym(nq,l,m,r,theta,phi):
    return R_sym(nq,l,r)*Ynm(l,m,theta,phi)

def pDens_Psi(nq,l,m,r,theta,phi):
    return (sympy.Abs(Psi_sym(3,1,0,r,theta,phi))**2)
```

```python
print("Showcasing the deconstruction of the position wavefunction:")
print("The radial wave function R(r)")
display(R_sym(2,0,r))
print("The spherical harmonic Y(theta,phi)")
display(Ynm(0,0,theta,phi).expand(func=True))
print("The wavefunction Psi(r,theta,phi) =  R(r)*Y(theta,phi), here the gound state, n = 2, m = 0, l = 0")
display(sympy.simplify(Psi_sym(2,0,0,r,theta,phi).expand(func=True)))
```

```python
display(R_sym(3,1,r))
display(sympy.simplify(Ynm(1,0,theta,phi).expand(func=True)))
display(sympy.simplify(Psi_sym(3,1,0,r,theta,phi).expand(func=True)))
display(sympy.simplify(pDens_Psi(3,1,0,r,theta,phi).expand(func=True)))
```

```python
print("The wavefunction Psi(r,theta,phi) =  R(r)*Y(theta,phi), n = 4, m = 2, l = 0")
display((1/sympy.sqrt(2))*(Psi_sym(2,1,1,r,theta,phi).expand(func=True)+Psi_sym(2,1,-1,r,theta,phi).expand(func=True)))
```

```python
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import sph_harm

phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2*np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

# The Cartesian coordinates of the unit sphere
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

m, l = 0, 4

# Calculate the spherical harmonic Y(l,m) and normalize to [0,1]
fcolors = sph_harm(m, l, theta, phi).real
fmax, fmin = fcolors.max(), fcolors.min()
fcolors = (fcolors - fmin)/(fmax - fmin)

# Set the aspect ratio to 1 so our sphere looks spherical
fig = plt.figure(figsize=plt.figaspect(1.))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z,  rstride=1, cstride=1, facecolors=cm.seismic(fcolors))
# Turn off the axis planes
ax.set_axis_off()
plt.show()
```
